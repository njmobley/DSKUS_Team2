```python
import pandas as pd

area = pd.read_csv('E:/석사 3기/DKSUS/area_9015_2.csv')
```

```python
area
```

```python
area_total = area['Item']=='Total'
```

```python
area[area_total]
```

```python
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np
```

```python
%matplotlib inline
plt.figure(figsize=(25,20))
lr = LinearRegression()

for i in range(1, len(area[area_total])):
    plt.subplot(4,3,i)
    x = np.array(range(1990, 2016, 1))
    y = area[area_total].iloc[i][3:].values
    plt.scatter(x,y)
    plt.title(area[area_total].iloc[i][0], fontsize=15)
    plt.xlabel("year", fontsize = 15)
    plt.ylabel("harvested area", fontsize=15)
    plt.xticks(fontsize=12)
    plt.yticks(fontsize=12)
    plt.grid()
    
    lr.fit(x.reshape(-1,1),y)
    plt.plot(x, lr.predict(x.reshape(-1,1)),c='c', linewidth=3)

plt.tight_layout()
```

```python
crop = pd.read_csv('E:/석사 3기/DKSUS/yield_9015_2.csv')
```

```python
crop_total = crop['Item']=='Total'
```

```python
crop[crop_total]
```

```python
%matplotlib inline
plt.figure(figsize=(25,20))
lr = LinearRegression()

for i in range(1, len(crop[crop_total])):
    plt.subplot(4,3,i)
    x = np.array(range(1990, 2016, 1))
    y = crop[crop_total].iloc[i][3:].values
    plt.scatter(x,y)
    plt.title(crop[crop_total].iloc[i][0], fontsize=15)
    plt.xlabel("year", fontsize = 15)
    plt.ylabel("crop yield", fontsize=15)
    plt.xticks(fontsize=12)
    plt.yticks(fontsize=12)
    plt.grid()
    
    lr.fit(x.reshape(-1,1),y)
    plt.plot(x, lr.predict(x.reshape(-1,1)),c='c', linewidth=3)

plt.tight_layout()
```

```python
from sklearn.preprocessing import PolynomialFeatures
%matplotlib inline
plt.figure(figsize=(25,20))
lr = LinearRegression()
poly = PolynomialFeatures(degree=5, include_bias=True)

for i in range(1,len(area[area_total])):
    plt.subplot(4,3,i)
    x = np.array(range(1990,2016,1)).reshape(-1,1)
    x_poly = poly.fit_transform(x)
    y = area[area_total].iloc[i][3:].values
    plt.scatter(x,y)
    plt.title(area[area_total].iloc[i][0], fontsize=15)
    plt.xlabel('year', fontsize=15)
    plt.ylabel("harvested area", fontsize=15)
    plt.xticks(fontsize=12)
    plt.yticks(fontsize=12)
    plt.grid()
    
    lr.fit(x_poly,y)
    plt.plot(x, lr.predict(x_poly), c = 'c', linewidth = 3)

plt.tight_layout()
```
