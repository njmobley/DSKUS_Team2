```python
c_yield = pd.read_csv('D:/석사 3기/DKSUS/yield_9015.csv')
c_yield.head()
```

```python
area.drop(['Domain Code','Domain','Area Code (FAO)','Element Code','Element','Item Code (FAO)','Year Code','Unit','Flag','Flag Description'], axis = 1, inplace=True)
```

```python
c_yield.drop(['Domain Code','Domain','Area Code (FAO)','Element Code','Element','Item Code (FAO)','Year Code','Unit','Flag','Flag Description'], axis = 1, inplace=True)
```

```python
area.info()
```

```python
c_yield.info()
```

```python
area_yield = pd.merge(area, c_yield, how='outer', on=['Area','Item','Year'])
```

```python
area_yield.info()
```

```python
area_yield = area_yield.rename(columns={'Value_x':'Value_area', 'Value_y':'Value_crop'})
```

```python
area_yield['Value_area'].fillna(area_yield.groupby(['Area','Item'])['Value_area'].transform('mean'), inplace=True)
```

```python
area_yield['Value_crop'].fillna(area_yield.groupby(['Area','Item'])['Value_crop'].transform('mean'), inplace=True)
```
